### Q2  	Draw a bar graph for number of matches won (i,e.consider win if the team wins both the toss and match)  for the team “Mumbai Indians” for all years
    

```python
import pandas as pd
import matplotlib.pyplot as plt
class tossMatch:
    def __init__(self):
        df=pd.read_csv('Match.csv')
    def ovw():
        val=df['Season_Year'][(df['match_winner']=='Mumbai Indians') & (df['Toss_Winner']=='Mumbai Indians')].value_counts()
        years=val.index.tolist()
        wins=val.tolist()
    def plot(years,wins):
        plt.title('Graph for number of matches won by “Mumbai Indians”')
        plt.xlabel('Wins')
        plt.ylabel('Years')
        plt.barh(years,wins,animated=True)
#         ax=df.plot.barh(y='years')
tm=tossMatch()
tm.ovw
# tm.wins
tossMatch.plot(years,wins)
```

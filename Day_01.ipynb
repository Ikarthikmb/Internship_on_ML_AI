### Q3 Calculate the average rating of Gaming and Non-Gaming apps

```python
games=[];other=[]
for i in range(len(y)):
    if y[i][12]=='Games':
        games.append(y[i][8])
    else:
        other.append(y[i][8])

o=len(games)
p=len(other)
print('Average rating of Gaming is',sum(games)/o)
print('Average rating of Non-Gaming is',sum(other)/p)
```

### Q7 Compute the average rating of the apps that have a price greater than $9.

### Q8 Categorise all apps by labelling each app as "free"(=0),  "affordable" (<20), "expensive" (<50) or "very expensive" (>50). Add a label column to the data.

```python
def check(l):
    if l==0:
        return 'Free'
    elif l<10:
        return 'Affordable'
    elif l<50:
        return 'Expensive'
    elif l>50:
        return 'Very Expensive'
```
